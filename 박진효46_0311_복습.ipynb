### 파생변수 추가하기
- 기존 데이터에 존재했던 변수에서 (mpg기준으로 보면 cty, hwy) 새로운 변수(mean_ch)를 추가하는 것
- assign 
- cty, hwy 의 평균 연비를 만들어보자! 
- 평균연비 만드는 식은 평균 구하는 식과 같다

```python
import numpy as np
```

```python
df = df.assign(mean_ch = (df['cty']+df['hwy'])/2)
# mean_ch 새롭게 만들 변수
```

```python
df.assign(grade = np.where(df['mean_ch']>= 20.15,'A','C'))
#np.where 넘파이에서 제공하는 Where함수는 해당 값의 위치를 찾아주는데 
```

```python
df['mean_ch'].mean()
```

```python
## lambda를 이용해서 만들기
df = df.assign(mean_ch1 = lambda x :(x['cty']+ x['hwy'])/2) #만들 수 있다.
```

```python
df['mean_ch2'] = (df['hwy']+df['cty'])/2 #정말 쉽게 
```

```python
np.where(df['mean_ch']< 20.15)
```

### Groupby
- 요약통계를 봐야할 때 수치형 데이터만 사용해서 볼 수 있는데
- 엑셀에서 피벗으로 돌려서 값을 만들 때 
- groupby(기준으로 묶을 컬럼), agg(값을 출력할 변수를 넣어주고, 원하는 통계치도 넣어준다) 함수와 같이 사용한다.

```python
df
```

```python
df.groupby('year').agg(mean_cty = ('cty','mean'), # ('변수', '내가 볼 통계치')
                     std_cty = ('cty','std'), #('cty','sum')
                     n = ('model','count')) #count는 개수 
```

```python
df['model']
```

```python
df.groupby('model').agg(mean_cty = ('cty','mean'), # ('변수', '내가 볼 통계치')
                     std_hwy = ('hwy','std'), #('cty','sum')
                     n = ('model','count')) #count는 개수 
```

```python
df
```

```python
df_2 =df.query("model =='a4'").groupby('year').agg(mean_cty = ('cty','mean'), # ('변수', '내가 볼 통계치')
                     sum_hwy = ('hwy','sum'), #('cty','sum')
                     n = ('model','count')) #count는 개수 
```

```python
df_2 #year는 인덱스다.
```

```python
df_2.columns # 컬럼이 3가지 밖에 없다.
```

```python
df_2 = df_2.reset_index() #인덱스를 리셋하는 방법 
```

### (데이터 합치기) merge  , concat
- 데이터프레임과 데이터프레임을 합친다.
- 왜?
- 데이터가 예쁘게 mpg처럼 있는 게 아니라 저장한 공간에 데이터가 나눠져 있을 건데, 테이블 별로
- 그 테이블에 있는 것들의 하나의 Mpg 데이터처럼 만들려면 합쳐야 된다.
- 특정상황에서- > 분석가 분석에 필요한 데이터를 합쳐야 하는 경우에 사용하면된다.
