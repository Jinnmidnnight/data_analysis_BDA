```python
import seaborn as sns
import numpy as np
```

```python
df_tt = sns.load_dataset('titanic')
```

```python
df_tt
```

### 데이터 컬럼 도메인 설명 ###
* survived: 0이면 사망, 1이면 생존
* pclass: 티켓 등급(1,2,3)
* sex/age: 성별/나이
* sibsp: 형제, 자매, 배우자의 합
* parch: 부모, 자식의 합
* fare: 요금
* embarked: 탑승 항구 C(Cherbourg), Q(Queenstown), S(Southampton)
* class: pclass와 동일
* who: 남자(man), 여자(woman), 아이(child)
* adult_male: 성인 남자 여부
* deck: 데크 번호 (알파벳 + 숫자 혼용)
* embark_town: 탑승 항구 이름
* alive: 생존여부 (yes, no)
* alone: 혼자 탑승 여부

```python
# '생존'이 중요한 키워드라, '생존'에 성공한 데이터를 y값에 assign
y = df_tt.query("survived == 1")
y
```

```python
# Q1: 어떠한 성별이 더 생존에 유리했는가
y.groupby('sex').agg(n = ('sex', 'count'))
```

```python
# Q2: 생존한 여성들에게 부모나 자식이 몇 명 타 있었는가
y.query("sex == 'female'").groupby('parch').agg(n = ('parch', 'count'))
```

```python
# Q3: 생존한 여성들에게 형제/자매나 배우자가 몇 명 타 있었는가
y.query("sex == 'female'").groupby('sibsp').agg(n = ('sibsp', 'count'))
```

```python
# Q3: 생존한 사람들의 좌석 등급 분포는 어떻게 되는가
y.groupby('pclass').agg(n = ('pclass', 'count'))
```

```python
# Q4: 생존한 남성/여성들의 평균 나이는 얼마인가
print(y.query("sex == 'male'")['age'].mean())
print(y.query("sex == 'female'")['age'].mean())
```

```python
# Q5: 생존한 사람들의 탑승 항구 분포가 어떻게 되는가
y.groupby('embark_town').agg(n = ('embark_town', 'count'))
```

```python
# Q6: 생존한 남성/여성들의 평균 요금은 얼마인가
print(y.query("sex == 'male'")['fare'].mean())
print(y.query("sex == 'female'")['fare'].mean())
```

```python
# Q7: 생존한 사람들 중 혼자 온 사람은 몇 명이나 되는가
y.groupby('alone').agg(n = ('alone', 'count'))
```

```python
# Q8: 생존한 사람들의 나이에 따른 요금 분포가 어떻게 되는가
y[['age','fare']].sort_values(['age','fare'], ascending=[False,False])
```

```python
# Q8: 생존한 사람들의 나이에 따른 부모/자식 수와 형제자매/배우자 수가 어떻게 되는가
y[['age','parch', 'sibsp']].sort_values(['age','parch','sibsp'], ascending=[True,False, False])
```

```python
# Q9: 생존한 남성 중 성인 남성은 얼마나 되는가
y.query("sex == 'male'").groupby('adult_male').agg(n = ('adult_male', 'count'))
```

```python
# Q9: 생존한 여성 중 성인 여성은 얼마나 되는가
y.query("sex == 'female'").groupby('who').agg(n = ('who', 'count'))
```

```python
# Q10: 성인 여성 요금은 좌석 등급에 따라 얼마나 되는가
y.query("who == 'woman'").groupby('pclass').agg(n = ('pclass', 'count'), fare = ('fare', 'mean'))
```
